```python
df['Credit Score']=df['Credit Score'].fillna(df['Credit Score'].mean())
#df['Years in current job']=df['Years in current job'].fillna(df['Years in current job'].mean())
df['Annual Income']=df['Annual Income'].fillna(df['Annual Income'].mean())
df['Months since last delinquent']=df['Months since last delinquent'].fillna(df['Months since last delinquent'].mean())
df['Bankruptcies']=df['Bankruptcies'].fillna(df['Bankruptcies'].mean())
df['Tax Liens']=df['Tax Liens'].fillna(df['Tax Liens'].mean())
```

```python
df["Years in current job"].mode() #ortalama olaak verinin rt aldı.
```

```python
df['Years in current job']=df['Years in current job'].fillna("10+ years")
```

```python
df.describe() # istatiksisel analiz değişkenlerin
```

```python
df.corr(numeric_only=True)
```

```python
df = df.drop_duplicates().copy()
```

```python
df["Purpose"].unique()
```

```python
df['Purpose'] = df['Purpose'].replace('Other', 'other')
```

```python
df["Purpose"].unique()
```

```python
df["Loan Status"].unique()
```

```python
df["Term"].unique()
```

```python
df["Home Ownership"].unique()
```

```python
df['Home Ownership'] = df['Home Ownership'].replace('HaveMortgage', 'Home Mortgage')
```

```python
df["Home Ownership"].unique()
```

```python

df["Credit Score"]=df["Credit Score"].astype("int32")#Selman İmdat
```

```python
df["Credit Score"].info()
```

```python
df['Monthly Debt'] = df['Monthly Debt'].str.replace(',',".") #virgülü noktaya çeviriyor.
```

```python

df['Monthly Debt'] = df['Monthly Debt'].str.replace('$',"").str.replace('.',"")
```

```python
df['Monthly Debt']=df['Monthly Debt'].fillna(0)
```

```python
df['Monthly Debt'] = df['Monthly Debt'].astype("int32")
```

```python
df['Monthly Debt'].info()
```

```python
df[['Monthly Debt']].head()
```

```python
df["Months since last delinquent"]=df["Months since last delinquent"].astype("int32") # integer a çevirdik çünkü object olarak bunu bilgsayara verfemeyiz
```

```python

df.sample(10)
```

```python
df['Maximum Open Credit'] = pd.to_numeric(df['Maximum Open Credit'], errors='coerce')
```

```python
df['Maximum Open Credit'] = df['Maximum Open Credit'].fillna(0).astype("int32")
```

```python
df["Loan Status"].isnull().sum()
```

```python
y=df["Loan Status"]
```

```python
x=df.drop(["Loan Status", "Customer ID", "Loan ID"] , axis=1)
```

```python
x=pd.get_dummies(x, drop_first=True) # x leri aldık sonrasında kategorikleri 1 0 a çevirdik.
```

```python
from sklearn.model_selection import train_test_split #şu kütüphaneden şunu çağırdım.
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB # yapay zeka modellerimiz
```

```python
g=GaussianNB()
b=BernoulliNB()
```

```python
g.fit(x_train, y_train)
```

```python
tahmin1=g.predict(x_test)
```

```python
b.fit(x_train, y_train)
```

```python
tahmin2=b.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(tahmin1, y_test)
```

```python
accuracy_score(tahmin2, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier # farklı bir model kullanacağız algoritma üretmek için
```

```python
k=KNeighborsClassifier()
```

```python
k.fit(x_train, y_train)
```

```python
tahmin3=k.predict(x_test)
```

```python
accuracy_score(tahmin3, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
l=LogisticRegression()
```

```python
l.fit(x_train, y_train)
```

```python
tahmin4=l.predict(x_test)
```

```python
accuracy_score(tahmin4, y_test)
```
